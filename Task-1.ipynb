```python
from keras.preprocessing.sequence import pad_sequences
from keras.models import load_model
import json
from tensorflow.keras.preprocessing.text import tokenizer_from_json
```

```python
def encode_sequences(tokenizer, length, lines):
    seq = tokenizer.texts_to_sequences(lines)
    seq = pad_sequences(seq, maxlen=length, padding='post')
    return seq
```

```python
model = load_model('/kaggle/input/french_to_english/keras/1/1/LSTM_model.h5')
```

```python
with open('/kaggle/input/french_to_english/keras/1/1/english_tokenizer.json') as f:
    data = json.load(f)
    english_tokenizer = tokenizer_from_json(data)
with open('/kaggle/input/french_to_english/keras/1/1/fra_tokenizer.json') as f:
    data = json.load(f)
    french_tokenizer = tokenizer_from_json(data)
```

```python
def get_word(n, tokenizer):
    for word, index in tokenizer.word_index.items():
        if index == n:
            return word
    return None
```

```python
sentence = 'je dois te montrer quelque chose'
```

```python
seq = encode_sequences(french_tokenizer,8, [sentence.lower()])[0]
```

```python
temp = []
for i in range(len(seq)):
    t = get_word(seq[i], english_tokenizer)
    if i > 0:
        if (t == get_word(seq[i-1], english_tokenizer)) or (t == None):
            temp.append('')
        else:
            temp.append(t)
    else:
        if(t == None):
            temp.append('')
        else:
            temp.append(t) 

print(' '.join(temp))
```
